```python
import numpy as np
from PIL import Image

def average_images(image_list):
    total_array = np.array(Image.open(image_list[0]), 'f')
    count=1

    for image_path in image_list[1:]:
        try:
            image_array = np.array(Image.open(image_path), 'f')
            total_array += image_array
            count=count+1
        except:
            print("Skip ", image_path);

    # Tnh trung bnh
    average_array = total_array / count

    # Chuyn i kt qu trung bnh thnh hnh nh
    average_image = Image.fromarray(average_array.astype('uint8'))

    return average_image

# S dng v d:
dir_1 = 'C:/Users/ADMIN/Documents/src_cv/img/01/1.jpg'
dir_2 = 'C:/Users/ADMIN/Documents/src_cv/img/01/2.jpg'
image_list = [dir_1, dir_2]
result_image = average_images(image_list)
display(result_image)


# S dng v d:
dir_1 = 'C:/Users/ADMIN/Documents/src_cv/img/01/3.jpg'
dir_2 = 'C:/Users/ADMIN/Documents/src_cv/img/01/4.jpg'
image_list = [dir_1, dir_2]
result_image = average_images(image_list)
display(result_image)
```
